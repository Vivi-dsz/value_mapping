```python
import pandas as pd

data=pd.read_csv('../data/raw/user_reviews_10k_v01.csv')
reviews=data['content']
reviews.head()
```

```python
##model for bank sentiment analysis (accuracy = 0.88)
from transformers import pipeline

sentiment_model = pipeline("text-classification", model="ajiayi/finetuning-sentiment-model-bank_reviews-otherbank-4insteadof3")

# Process all reviews
results = [sentiment_model(r)[0] for r in reviews]

# Combine results into DataFrame
df_sentiment = pd.DataFrame({
    "review": reviews,
    "sentiment": [r["label"] for r in results],
    "confidence": [r["score"] for r in results]
})

df_sentiment.head()
```
